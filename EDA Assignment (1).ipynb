## Scaling
Scaling methods are used to normalize the range of independent variables or features of data. Here are some of the most common scaling methods:

Standard scaler

Mainly used this method
formulla-xi-xbar/sigma

Min-Max Scaling (Normalization)

Scales the data to a fixed range, usually 0 to 1.
Formula: X′=Xmax−XminX−Xmin

Robust Scaler

Scales the data according to the interquartile range.
Useful for data with many outliers.
Formula: X′=IQRX−median
IQR is the Interquartile Range (75th percentile - 25th percentile).
